```python
import pandas as pd
import numpy as np
import os
```

```python
os.chdir(os.path.dirname(os.getcwd()))
```

```python
df = pd.read_csv("data/raw/googleplaystore.csv", index_col=0)
```

```python
df
```

```python
df.loc[df["Reviews"]==0,"Rating"]=0.0
```

```python
df.sort_values(by="App", inplace=True)
df.reset_index(inplace=True,drop=True)
```

```python
for i in range(0,len(df)):
    if i==0:
        continue
    if df.loc[i,"App"]==df.loc[(i-1),"App"] and df.loc[i,"Category"]==df.loc[(i-1),"Category"]:
        df.loc[i,"duplicado"]=1
    else:
        df.loc[i,"duplicado"]=0
```

```python
df.drop(df[df["duplicado"]==1].index,axis=0, inplace=True)
```

```python
df.info()
```

```python
df.drop("duplicado",axis=1,inplace=True)
```

```python
df.info()
```

```python
df.drop(df[df["Rating"].isna()].index,axis=0, inplace=True)
```

```python
df["Rating"] = np.where(df["Rating"].str.contains(","),df["Rating"].str.replace(",","."),df["Rating"])
df.loc[df["Rating"].isna(),"Rating"]=0
df["Rating"]=df["Rating"].astype(float)
df.info()
```

```python
df
```

```python
mediana = pd.DataFrame(df.groupby('Category')['Rating'].median().sort_values(ascending=True))
```

```python
for i,j in enumerate(list(df.groupby('Category')['Rating'].median().sort_values(ascending=True).index)):
    if mediana.loc[j].values[0]==4.5:
        df.loc[df["Category"]==j,"medianacategoria"]=4
    if mediana.loc[j].values[0]==4.4:
        df.loc[df["Category"]==j,"medianacategoria"]=3
    elif mediana.loc[j].values[0]==4.3:
        df.loc[df["Category"]==j,"medianacategoria"]=2
    elif mediana.loc[j].values[0]==4.2:
        df.loc[df["Category"]==j,"medianacategoria"]=1
    elif mediana.loc[j].values[0]==4.1:
        df.loc[df["Category"]==j,"medianacategoria"]=0
```

```python
for i, j in enumerate(list(df.groupby('Category')['Rating'].mean().sort_values(ascending=True).index)):
    df.loc[df["Category"]==j,"ncategoria"]=i
```

```python
for i in range(0,len(df["Reviews"])):
    if df.loc[i,"Reviews"]==0:
        df.loc[i,"valoraciones"]=0
    elif df.loc[i,"Reviews"]>0 and df.loc[i,"Reviews"]<=10:
        df.loc[i,"valoraciones"]=1
    elif df.loc[i,"Reviews"]>10 and df.loc[i,"Reviews"]<=100:
        df.loc[i,"valoraciones"]=2
    elif df.loc[i,"Reviews"]>100 and df.loc[i,"Reviews"]<=1000:
        df.loc[i,"valoraciones"]=3
    elif df.loc[i,"Reviews"]>1000 and df.loc[i,"Reviews"]<=10000:
        df.loc[i,"valoraciones"]=4
    elif df.loc[i,"Reviews"]>10000 and df.loc[i,"Reviews"]<=100000:
        df.loc[i,"valoraciones"]=5
    elif df.loc[i,"Reviews"]>100000:
        df.loc[i,"valoraciones"]=6
```
